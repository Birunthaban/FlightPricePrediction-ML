Change Duration into one unit (minutes)

```python
def changeDurationFormat(duration:str):
  result =duration.split()
  if len(result)==2:
    hours =int(duration.split()[0][0:-1])
    minutes = int(duration.split()[1][0:-1])
    return hours*60 + minutes
  else:
    if 'h' in result[0]:
      hours=int(result[0][0:-1])*60
      return hours
    else:
      minutes =int(result[0][0:-1])
      return minutes
# def changeDurationFormat2(duration:str):
# return eval(duration.replace("h","*60").replace("m","*1").replace(" ","+"))

```

```python
def cleanDuration(data:pd.DataFrame):
  duration =data['Duration']
  return duration.apply(changeDurationFormat)
```

```python
X_train['Duration'] = cleanDuration(X_train)
```

```python
def extractDepTime(data:pd.DataFrame):
  data['Dep_Time_H']=data['Dep_Time'].apply(lambda x :int(x.split(":")[0]))
  data['Dep_Time_M']=data['Dep_Time'].apply(lambda x :int(x.split(":")[1]))
  data.drop('Dep_Time',axis=1,inplace=True)


```

```python
extractDepTime(X_train)
extractDepTime(X_test)
```

```python
def extractDateFeatures(data:pd.DataFrame):
  data['Journey_Day']=data['Date_of_Journey'].apply(lambda x :int((str(x).split("/")[0])))
  data['Journey_Month']=data['Date_of_Journey'].apply(lambda x :int(str(x).split("/")[1]))
  data['Dep_Time_Year']=data['Date_of_Journey'].apply(lambda x :int(str(x).split("/")[2]))
  data.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
extractDateFeatures(X_train)
extractDateFeatures(X_test)
```

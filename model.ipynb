# Check Duplicates

##### Duplicates can lead to bias in the model's training process. If certain records are duplicated more than others, the model might become biased towards the duplicated instances, potentially leading to overfitting. Overfitting occurs when a model learns noise in the data rather than the underlying patterns, resulting in poor generalization to new, unseen data.
